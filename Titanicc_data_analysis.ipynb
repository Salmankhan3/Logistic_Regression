```python
titanic_data.head()
```

```python
sex=pd.get_dummies(titanic_data['sex'],drop_first=True)
sex=sex.astype(int)
sex
```

```python
pc=pd.get_dummies(titanic_data['pclass'],drop_first=True)
pc=pc.astype(int)
pc
```

```python
embarke=pd.get_dummies(titanic_data['embarked'],drop_first=True)
embarke=embarke.astype(int)
embarke
```

```python
titanic_data=pd.concat([titanic_data,sex,pc,embarke],axis=1)
```

```python
titanic_data.head(4)
```

```python
titanic_data.drop(['name','cabin','boat','home.dest','sex','pclass','embarked'],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.drop('ticket',axis=1,inplace=True)
```

```python
titanic_data
```

```python
x=titanic_data.drop('survived',axis=1)
y=titanic_data['survived']
y,x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
X_train.columns = X_train.columns.astype(str)
X_train
```

```python
logmodel.fit(X_train,Y_train)
```

```python
X_test.columns=X_test.columns.astype(str)

prediction=logmodel.predict(X_test)
```

```python
X_test
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(Y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
labels=[0,1]
confusion_matrix(Y_test,prediction,labels=labels)
```
